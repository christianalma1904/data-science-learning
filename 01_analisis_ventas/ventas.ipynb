## Analisis de Ventas

```python
#fecha,producto,categoria,precio,cantidad,ciudad
#2025-01-01,Audifonos,Electronica,25.5,2,Quito
#2025-01-01,Mouse,Electronica,15.0,1,Guayaquil
#2025-01-02,Camiseta,Ropa,10.0,3,Quito
#2025-01-02,Pantalon,Ropa,20.0,1,Cuenca
#2025-01-03,Audifonos,Electronica,25.5,1,Quito
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'fecha':['2025-01-01','2025-01-01','2025-01-02','2025-01-02','2025-01-03'],
    'producto':['Audifonos','Mouse','Camiseta','Pantalon','Audifonos'],
    'categoria':['Electronica','Electronica','Ropa','Ropa','Electronica'],
    'precio':[25.5,15.0,10.0,20.0,25.5],
    'cantidad':[2,1,3,1,1],
    'ciudad':['Quito','Guayaquil','Quito','Cuenca','Quito']
})

df['fecha'] = pd.to_datetime(df['fecha']) #ojo
df
```

```python
df[0:2]
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['total_venta'] = df['precio'] * df['cantidad']
df
```

```python
df.rename(columns={'fecha': 'tipo_fecha'}, inplace=True)
df
```

```python
df1 = df['ciudad'] == 'Quito'
df[df1]
```

```python
df2 = df['categoria'] == 'Electronica'
df[df2]
```

```python
df3 = df['total_venta'].sum()
df3
```

```python
ingreso_por_ciudad = df.groupby('ciudad')['total_venta'].sum()
ingreso_por_ciudad
```

```python
ventas_producto = df.groupby('producto')['cantidad'].sum().sort_values(ascending=False)
ventas_producto
```

```python
precio_promedio = df.groupby('categoria')['precio'].mean()
precio_promedio
```

```python
tventas = df['total_venta'].to_numpy()
promedio = np.mean(tventas)
maximo = np.max(tventas)
desviacion = np.std(tventas, ddof=1)

print(f"Promedio: {promedio}")
print(f"Máximo: {maximo}")
print(f"Desviación estándar: {desviacion}")
```
